```python
df['Engine Cylinders'] = df['Engine Cylinders'].fillna(4.0)
```

```python
newnew_df = df.loc[df['Make'] == 'Lotus']
HP_cylinders = newnew_df[["Engine HP", "Engine Cylinders"]]
HP_cylinders.drop_duplicates(inplace=True)
HP_cylinders.shape[0]
```

```python
x = np.array(HP_cylinders)
xtx = np.dot(x.T,x)
xtxinv = np.linalg.inv(xtx)


```

```python
y=np.array([1100, 800, 750, 850, 1300, 1000, 1000, 1300, 800])
w = np.dot(np.dot(xtxinv,x.T),y)
w
```
